```python
population_by_country_2020 = population_by_country_2020.loc[:,["Country (or dependency)","Population (2020)"]]

population_by_country_2020.head()
```

```python
# change columes name
population_by_country_2020 = population_by_country_2020.rename(columns={"Country (or dependency)":"Country_NAME","Population (2020)":"Population"})


# # change China to Mainland China, United States to US
population_by_country_2020.at[0,"Country_NAME"]="Mainland China"
population_by_country_2020.at[2,"Country_NAME"]="US"

population_by_country_2020.head()
```

```python
import json

with open("./countries.json","r") as file:
    json_arr = json.load(file)
    
#swap key value
json_arr_s = {json_arr[key]:key for key in json_arr}
```

```python
def transformFlag(CountryName):
    try:
        url = "/flags/"+json_arr_s[CountryName].lower()+".png"
        return url
    except:
        return "/flags/default.png"

population_by_country_2020["flag_url"] = population_by_country_2020["Country_NAME"]. \
                        apply(transformFlag)
    
population_by_country_2020.head()
```

```python
population_by_country_2020.to_csv("./population_clean.csv",index=False)
```

```python
import json
with open("countries.geo.json") as f:
    data = json.load(f)
    
Geo_Countries = []
for country in data["features"]:
    Geo_Countries.append(country["properties"]["name"])
```

```python
def addGeo(country_name):
    for geo in Geo_Countries:
        if(geo == country_name):
            return geo
    return ""
```

```python
population_by_country_2020.head()
```

```python
population_by_country_2020["Geo"] = population_by_country_2020["Country_NAME"].apply(lambda x:addGeo(x))
population_by_country_2020.head()
```

```python
population_by_country_2020.to_csv("./population_clean_geo.csv",index=False)
```

```python
import pandas as pd

covid_19_data = pd.read_csv("./covid_19_data.csv")

covid_19_data = covid_19_data.drop(["Last Update"],axis=1)
covid_19_data = covid_19_data.drop(["SNo"],axis=1)

covid_19_data.head()
```

```python
# i = 0
total_row = [["ObervationDate","Country","Confirmed","Deaths","Recovered"]]
date_threshold = "01/22/2020"
oneDay_row = {}

for index,row in covid_19_data.iterrows():
    date = row["ObservationDate"]
    country = row["Country/Region"]
    
    if(date_threshold != date):
        for c in oneDay_row:
            total_row.append([date_threshold,c]+oneDay_row[c])
        oneDay_row = {}
        date_threshold = date
        
    
    if(country in oneDay_row):
        oneDay_row[country][0] += int(row["Confirmed"])
        oneDay_row[country][1] += int(row["Deaths"])
        oneDay_row[country][2] += int(row["Recovered"])
    else:
        oneDay_row[country] = [int(row["Confirmed"]),int(row["Deaths"]),int(row["Recovered"])]
    
#     if(i>100): break
#     i +=1

for c in oneDay_row:
    total_row.append([date_threshold,c]+oneDay_row[c])
```

```python
import csv 

with open("Country_covid.csv","w",newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(total_row)
```

```python
import csv
import pandas as pd
import json
```

```python
df = pd.read_csv("./time_series_covid_19_confirmed_US.csv")
df.head()
```

```python
df.describe(include="all")
```

```python
csv_confirmed = []
with open("./time_series_covid_19_confirmed_US.csv") as csvfile:
    data = csv.reader(csvfile)
    for d in data:
        csv_confirmed.append(d)

header = csv_confirmed[0]
csv_confirmed = csv_confirmed[1:]
    
csv_deaths = []
with open("./time_series_covid_19_deaths_US.csv") as csvfile:
    data = csv.reader(csvfile)
    for d in data:
        csv_deaths.append(d)
csv_deaths = csv_deaths[1:]
```

```python
print(len(csv_confirmed))
```

```python
state = set()
for d in csv_confirmed:
    state.add(d[6])
print(list(state))
print(len(list(state)))
```

```python
with open("us-states.json") as f:
    data = json.load(f)
state_geo = []
for feature in data["features"]:
    state_geo.append(feature["properties"]["name"])
print(state_geo)
print(len(state_geo))
```

```python
state_con = {}
for s in csv_confirmed:
    state_name = s[6]
    if(state_name in state_con):
        for i in range(11,len(s)):
            state_con[state_name][i-11] += int(s[i])
    else:
        state_con[state_name] = [int(s[i]) for i in range(11,len(s))]
```

```python
ls = []
for key in state_con:
    if(key not in state_geo):
        ls.append(key)
for key in ls:
    del state_con[key]
```

```python
state_deaths = {}
state_population = {}
for s in csv_deaths:
    state_name = s[6]
    if(state_name in state_deaths):
        for i in range(12,len(s)):
            state_deaths[state_name][i-12] += int(s[i])
        state_population[state_name] += int(s[11])
    else:
        state_deaths[state_name] = [int(s[i]) for i in range(12,len(s))]
        state_population[state_name] = int(s[11])
```

```python
ls = []
for key in state_deaths:
    if(key not in state_geo):
        ls.append(key)
for key in ls:
    del state_deaths[key]
    del state_population[key]
```

```python
print(state_population)
print(len(state_population))
```

```python
states_lower = ["-".join(geo.lower().split(" ")) for geo in state_geo]

import os
imgs = os.listdir("./all-flags/")
imgs_have = []
for img in imgs:
    if(img.lower().split(".")[0] in states_lower):
        imgs_have.append(img)

print(imgs_have)
print(len(imgs_have))
```

```python
state_csv = []
state_csv.append(["state_name","population","flag_url","geo"])
for state in state_population:
    flag_url = "/US_flag/"+"-".join(state.lower().split(" "))+".svg"
    arr = [state,state_population[state],flag_url,state]
    state_csv.append(arr)
```

```python
with open("./US_state.csv","w",encoding="utf8",newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(state_csv)
```

```python
state_country = []
state_country.append(["country_name","state_name"])
for state in state_geo:
    arr = ["US",state]
    state_country.append(arr)
    
with open("./country_state.csv","w",encoding="utf8",newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(state_country)
```

```python
total_covid = []
total_covid.append(["country_name","state_name","s_date","confirmed","deaths"])
for index,date in enumerate(header[11:]):
    for state in state_con:
        confirmed = state_con[state][index]
        deaths = state_deaths[state][index]
        if(confirmed == 0 and deaths == 0):
            continue
        arr = ["US",state,date,state_con[state][index],state_deaths[state][index]]
        total_covid.append(arr)
```
