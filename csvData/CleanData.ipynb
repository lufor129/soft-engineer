## covid_19_data.csv

```python
import pandas as pd
```

```python
covid_19_data = pd.read_csv("./covid_19_data.csv")

covid_19_data.describe(include="all")
```

```python
# isNULL
print(pd.isnull(covid_19_data).sum())
```

```python
covid_19_data = covid_19_data.drop(["Last Update"],axis=1)
covid_19_data = covid_19_data.drop(["SNo"],axis=1)

covid_19_data.head()
```

```python
covid_19_data["Confirmed"] = covid_19_data["Confirmed"].apply(lambda x:int(x))
covid_19_data["Deaths"] = covid_19_data["Deaths"].apply(lambda x:int(x))
covid_19_data["Recovered"] = covid_19_data["Recovered"].apply(lambda x:int(x))
```

```python
# change columes name and save
covid_19_data = covid_19_data.rename(columns={"ObservationDate":"CC_date",
                                              "Province/State":"Province_NAME",
                                              "Country/Region":"Country_NAME"
                                             })
covid_19_data.head()
```

```python
covid_19_data.to_csv("./covid_clean.csv",index=False)
```

```python
population_by_country_2020 = pd.read_csv("./population_by_country_2020.csv")

population_by_country_2020.describe(include="all")
```

```python
print(pd.isnull(population_by_country_2020).sum())
```

```python
population_by_country_2020.head()
```

```python
population_by_country_2020 = population_by_country_2020.loc[:,["Country (or dependency)","Population (2020)"]]

population_by_country_2020.head()
```

```python
# change columes name
population_by_country_2020 = population_by_country_2020.rename(columns={"Country (or dependency)":"Country_NAME","Population (2020)":"Population"})


# # change China to Mainland China, United States to US
population_by_country_2020.at[0,"Country_NAME"]="Mainland China"
population_by_country_2020.at[2,"Country_NAME"]="US"

population_by_country_2020.head()
```

```python
import json

with open("./countries.json","r") as file:
    json_arr = json.load(file)
    
#swap key value
json_arr_s = {json_arr[key]:key for key in json_arr}
```

```python
def transformFlag(CountryName):
    try:
        url = "/flags/"+json_arr_s[CountryName].lower()+".png"
        return url
    except:
        return "/flags/default.png"

population_by_country_2020["flag_url"] = population_by_country_2020["Country_NAME"]. \
                        apply(transformFlag)
    
population_by_country_2020.head()
```

```python
population_by_country_2020.to_csv("./population_clean.csv",index=False)
```

```python
import json
with open("countries.geo.json") as f:
    data = json.load(f)
    
Geo_Countries = []
for country in data["features"]:
    Geo_Countries.append(country["properties"]["name"])
```

```python
def addGeo(country_name):
    for geo in Geo_Countries:
        if(geo == country_name):
            return geo
    return ""
```

```python
population_by_country_2020.head()
```

```python
population_by_country_2020["Geo"] = population_by_country_2020["Country_NAME"].apply(lambda x:addGeo(x))
population_by_country_2020.head()
```

```python
population_by_country_2020.to_csv("./population_clean_geo.csv",index=False)
```

```python
import pandas as pd

covid_19_data = pd.read_csv("./covid_19_data.csv")

covid_19_data = covid_19_data.drop(["Last Update"],axis=1)
covid_19_data = covid_19_data.drop(["SNo"],axis=1)

covid_19_data.head()
```

```python
# i = 0
total_row = [["ObervationDate","Country","Confirmed","Deaths","Recovered"]]
date_threshold = "01/22/2020"
oneDay_row = {}

for index,row in covid_19_data.iterrows():
    date = row["ObservationDate"]
    country = row["Country/Region"]
    
    if(date_threshold != date):
        for c in oneDay_row:
            total_row.append([date_threshold,c]+oneDay_row[c])
        oneDay_row = {}
        date_threshold = date
        
    
    if(country in oneDay_row):
        oneDay_row[country][0] += int(row["Confirmed"])
        oneDay_row[country][1] += int(row["Deaths"])
        oneDay_row[country][2] += int(row["Recovered"])
    else:
        oneDay_row[country] = [int(row["Confirmed"]),int(row["Deaths"]),int(row["Recovered"])]
    
#     if(i>100): break
#     i +=1

for c in oneDay_row:
    total_row.append([date_threshold,c]+oneDay_row[c])
```

```python
print(total_row[:40])
```

```python
import csv 

with open("Country_covid.csv","w",newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(total_row)
```
